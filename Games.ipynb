```python
# Blackjack

import random

score = 0
player_score = 0
x = ""

def hit():
  global score
  x = random.randint(1, 13)

  if x >= 10:
    score += 10

    if x == 11:
      return "J"
    elif x == 12:
      return "Q"
    elif x == 13:
      return "K"
    else:
      return x

  else:
    score += x

    if x == 1:
      return "A"
    else:
      return x


def check_lose():
  if score >= 22:
    return "l"
  else:
    return "x"

# first hit before loop
print("Hit:", hit())

# loop until player stands or loses
while x.lower() != "stand" and check_lose() != "l":
  print("Hit:", hit())
  if check_lose() == "l":
    print("Bust")
    print("Dealer Wins")
  else:
    x = input("Hit or Stand: ")

if x.lower() == "stand" and check_lose() != "l":
  player_score = score
  score = 0

  while check_lose() != "l" and score <= 17:
    print("Dealer Hit:", hit())

  if check_lose() == "l":
      print("Dealer Bust")
      print("Player Wins")
  else:
    print("Player Score:", player_score, "  Dealer Score:", score)
    if score > player_score:
      print("Dealer Wins")
    elif player_score > score:
      print("Player Wins")
    else:
      print("Tie")
```

```python
# Chess

from tabulate import tabulate



def start_pos():
  global board
  board = [
    ["r","n","b","q","k","b","n","r"],
    ["p","p","p","p","p","p","p","p"],
    ["","","","","","","",""],
    ["","","","","","","",""],
    ["","","","","","","",""],
    ["","","","","","","",""],
    ["P","P","P","P","P","P","P","P"],
    ["R","N","B","Q","K","B","N","R"]
  ]


def display_board():
  print("\n" + tabulate(board, tablefmt="grid") + "\n")


def move(x, y):
  x = coord(x)
  y = coord(y)

  if check_move(x, y, board[x[1]][x[0]]):
    board[y[1]][y[0]] = board[x[1]][x[0]]
    board[x[1]][x[0]] = ""


def check_move(x, y, z):

  # PAWNS
  if z == "p":
    # one square down, one square either side, moving to white occupied space
    if y[1] == x[1] + 1 and (y[0] - x[0])**2 == 1 and board[y[1]][y[0]].isupper():
      return True
    # on home row, two squares down, file is same, moving to empty square
    elif x[1] == 1 and y[1] == x[1] + 2 and y[0] == x[0] and board[y[1]][y[0]] == "":
      return True
    # one square down, file is same, moving to empty square
    elif y[1] == x[1] + 1 and y[0] == x[0] and board[y[1]][y[0]] == "":
      return True
    return False

  elif z == "P":
    # one square up, one square either side, moving to black occupied space
    if y[1] == x[1] - 1 and (y[0] - x[0])**2 == 1 and board[y[1]][y[0]].islower():
      return True
    # on home row, two squares up, file is same, moving to empty square
    elif x[1] == 6 and y[1] == x[1] - 2 and y[0] == x[0] and board[y[1]][y[0]] == "":
      return True
    # one square up, file is same, moving to empty square
    elif y[1] == x[1] - 1 and y[0] == x[0] and board[y[1]][y[0]] == "":
      return True
    return False

  # ROOKS
  if z.lower() == "r":
    # end if neither file nor row is same, those are all diagonals
    if y[1] != x[1] and y[0] != x[0]:
      return False
    # end if square occupied by same colour
    if z.islower() and board[y[1]][y[0]].islower():
      return False
    if z.isupper() and board[y[1]][y[0]].isupper():
      return False
    # up
    if y[1] < x[1]:
      for i in range(1, abs(y[1] - x[1]), 1):
          if board[x[1] - i][x[0]] != "":
            return False
    # down
    if y[1] > x[1]:
      for i in range(1, abs(y[1] - x[1]), 1):
          if board[x[1] + i][x[0]] != "":
            return False
    # right
    if y[0] > x[0]:
      for i in range(1, abs(y[0] - x[0]), 1):
          if board[x[1]][x[0] + i] != "":
            return False
    # left
    if y[0] < x[0]:
      for i in range(1, abs(y[0] - x[0]), 1):
          if board[x[1]][x[0] - i] != "":
            return False
    # return true if all checks pass
    return True

  # KNIGHTS
  elif z == "n":
    # either dx=2 and dy=1 OR dx=1 and dy=2
    if ((abs(y[1] - x[1]) == 2 and abs(y[0] - x[0]) == 1) or (abs(y[1] - x[1]) == 1 and abs(y[0] - x[0]) == 2)) and not board[y[1]][y[0]].islower():
      return True
    return False

  elif z == "N":
    # either dx=2 and dy=1 OR dx=1 and dy=2
    if ((abs(y[1] - x[1]) == 2 and abs(y[0] - x[0]) == 1) or (abs(y[1] - x[1]) == 1 and abs(y[0] - x[0]) == 2)) and not board[y[1]][y[0]].isupper():
      return True
    return False

  # BISHOPS
  if z.lower() == "b":
    # end if not on exact diagonal
    if abs(y[1] - x[1]) != abs(y[0] - x[0]):
      return False
    # end if square occupied by same colour
    if z.islower() and board[y[1]][y[0]].islower():
      return False
    if z.isupper() and board[y[1]][y[0]].isupper():
      return False
    # up-right
    if y[1] < x[1] and y[0] > x[0]:
      for i in range(1, abs(y[1] - x[1]), 1):
        if board[x[1] - i][x[0] + i] != "":
          return False
    # down-right
    if y[1] > x[1] and y[0] > x[0]:
      for i in range(1, abs(y[1] - x[1]), 1):
        if board[x[1] + i][x[0] + i] != "":
          return False
    # down-left
    if y[1] > x[1] and y[0] < x[0]:
      for i in range(1, abs(y[1] - x[1]), 1):
        if board[x[1] + i][x[0] - i] != "":
          return False
    # up-left
    if y[1] < x[1] and y[0] < x[0]:
      for i in range(1, abs(y[1] - x[1]), 1):
        if board[x[1] - i][x[0] - i] != "":
          return False
    # return true if all checks pass
    return True

  # QUEENS
  elif z.lower() == "q":
    # end if square occupied by same colour
    if z.islower() and board[y[1]][y[0]].islower():
      return False
    if z.isupper() and board[y[1]][y[0]].isupper():
      return False
    # acting as bishop or rook?
    if (y[1] == x[1]) ^ (y[0] == x[0]):
      # up
      if y[1] < x[1]:
        for i in range(1, abs(y[1] - x[1]), 1):
            if board[x[1] - i][x[0]] != "":
              return False
      # down
      if y[1] > x[1]:
        for i in range(1, abs(y[1] - x[1]), 1):
            if board[x[1] + i][x[0]] != "":
              return False
      # right
      if y[0] > x[0]:
        for i in range(1, abs(y[0] - x[0]), 1):
            if board[x[1]][x[0] + i] != "":
              return False
      # left
      if y[0] < x[0]:
        for i in range(1, abs(y[0] - x[0]), 1):
            if board[x[1]][x[0] - i] != "":
              return False
      # return true if all checks pass
      return True
    elif abs(y[1] - x[1]) == abs(y[0] - x[0]):
      # up-right
      if y[1] < x[1] and y[0] > x[0]:
        for i in range(1, abs(y[1] - x[1]), 1):
          if board[x[1] - i][x[0] + i] != "":
            return False
      # down-right
      if y[1] > x[1] and y[0] > x[0]:
        for i in range(1, abs(y[1] - x[1]), 1):
          if board[x[1] + i][x[0] + i] != "":
            return False
      # down-left
      if y[1] > x[1] and y[0] < x[0]:
        for i in range(1, abs(y[1] - x[1]), 1):
          if board[x[1] + i][x[0] - i] != "":
            return False
      # up-left
      if y[1] < x[1] and y[0] < x[0]:
        for i in range(1, abs(y[1] - x[1]), 1):
          if board[x[1] - i][x[0] - i] != "":
            return False
      # return true if all checks pass
      return True
    return False

  # KINGS
  elif z == "k":
    # if one square in either direction, not black occupied space:
    if (y[1] - x[1])**2 <= 1 and (y[0] - x[0])**2 <= 1 and not board[y[1]][y[0]].islower():
      return True
    return False

  elif z == "K":
    # if one square in either direction, not white occupied space:
    if (y[1] - x[1])**2 <= 1 and (y[0] - x[0])**2 <= 1 and not board[y[1]][y[0]].isupper():
      return True
    return False

  # EMPTY
  else: return False


def coord(x):
  y = [x[0], int(x[1])]

  if y[0] == "A":
    y[0] = 0
  elif y[0] == "B":
    y[0] = 1
  elif y[0] == "C":
    y[0] = 2
  elif y[0] == "D":
    y[0] = 3
  elif y[0] == "E":
    y[0] = 4
  elif y[0] == "F":
    y[0] = 5
  elif y[0] == "G":
    y[0] = 6
  elif y[0] == "H":
    y[0] = 7

  y[1] = 8 - y[1]

  return y


def check_win():
  white_win = True
  black_win = True

  for i in range(len(board)):
    if "k" in board[i]:
      white_win = False
    if "K" in board[i]:
      black_win = False

  if white_win == True:
    return "white"
  elif black_win == True:
    return "black"
  else: return ""



start_pos()
display_board()

while check_win() == "":
  move(input("Which piece to move? "), input("Where to? "))
  display_board()

if check_win() == "white":
  print("*** WHITE WINS ***")

if check_win() == "black":
  print("*** BLACK WINS ***")

print("test git")
```
